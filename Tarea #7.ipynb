```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fun_gap(E):
    h = 6.58211e-16
    m = 9.109382e-31
    alpha = 1
    a = 1.5
    b = (m*a*alpha)/(h**2)
    k = np.sqrt(2*m*E/(h**2))
    fun = np.cos(k*a) +(b * (np.sin(k*a)/k*a))
    return fun 
```

```python
def cos_a(l):
    a = 1.5
    N = 100
    phi = (2*np.pi*l)/(N*a)
    fun = np.cos(phi*a)
    return fun
    
```

```python
l = np.arange(101)
E = np.linspace(0.001,20,10000)
bandas = np.zeros((N+1,len(E)))
j = fun_gap(E)

for i in range(len(l)):
    for z in range(len(E)):
        if j[z] <= 1 and j[z] >= -1:
            bandas[i,z] = cos_a(l[i])
        else:
            bandas[i,z] = None
```

```python
plt.figure(figsize = (15,13))
for n in range(len(l)):
    plt.plot(E, bandas[n], color = 'g')
plt.plot(E , fun_gap(E)) 
plt.xlabel("Energia")
plt.ylabel("banda")
plt.title("bandas de conduccion")
plt.show()
```
